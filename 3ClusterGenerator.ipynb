```
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'

import numpy as np
import math

from google.colab import drive

drive.mount("/content/gdrive", force_remount=True)
```

```
PATH_TO_CSV = "gdrive/My Drive/Machine_learning_team/Assignment_2/Data/initial_dataset.csv"
df = pd.read_csv(PATH_TO_CSV)
df.head()
```

```
# Ordenar las minucias de acuerdo al score_change
df.sort_values(by=['score_change'], ascending=False, inplace=True)
```

```
df.reset_index(inplace=True)
```

```
df.drop(columns=['index'], inplace=True)

subset_df = df[df == '?']

column_count = subset_df.count()
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#    print(column_count)

toMedian = []

for i, v in column_count.iteritems():
  if v > 0:
    toMedian.append(i)

print(toMedian)
```

```
df.replace('?', np.NaN, inplace=True)
median_dict = {}
for c in toMedian:
  median = df[c].median(skipna=True)
  median_dict[c] = median

print(median_dict)
```

```
for c in toMedian:
  df[c].replace(np.NaN, median_dict[c], inplace=True)
```

```
import random
random.seed(1)
```

```
# Hacer los 50 datasets

PATH_TO_SAVE = "gdrive/My Drive/Machine_learning_team/Assignment_2/Data/3CLUSTER/"
FILE = "3CLUSTER_"
EXT = ".csv"
for i in range(0, 50):
  # resetear la columna class
  df['class'] = ''
  p1 = random.randint(29, 5175)
  p2 = random.randint(p1+30, 5210)
  df['class'][:p1] = '+'
  df['class'][p1:p2] = '0'
  df['class'][p2:] = '-'

  df.to_csv(PATH_TO_SAVE + FILE + str(i) + EXT)

  print(df.groupby('class').count())
```
